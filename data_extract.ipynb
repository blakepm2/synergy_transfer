```python
import pandas as pd
import json
from alive_progress import alive_bar
```

```python
summary = pd.read_csv('data1.4/summary_v_1_5.csv')
```

```python
with open('data1.4/studies.json', 'r') as infile:
    studies = json.load(infile)
    infile.close()
```

```python
studies[0]
```

```python
studies_ = {}
with alive_bar(len(studies)) as bar:
    for i, s in enumerate(studies):
        for k in s.keys():
            if k not in studies_:
                studies_[k] = [s[k]]
            else:
                studies_[k].append(s[k])
        bar.text(f'Processing Study ({i+1}/{len(studies)})')
        bar()
```

```python
studies_df = pd.DataFrame(studies_)
studies_df.rename(columns={'sname': 'study_name', 'id': 'study_id'}, inplace=True)
```

```python
studies_df
```

```python
studies_df.to_csv('data1.4/studies.csv', index=False)
```

```python
summary_study=pd.merge(summary, studies_df[['study_name', 'study_id']], how='left',left_on='study_name', right_on='study_name')
```
